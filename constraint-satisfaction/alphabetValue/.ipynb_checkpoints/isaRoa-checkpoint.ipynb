### CRYPTARITHMETIQUE

Problem: ISA + ROA = TELO

Modelization:

- Variables: {I, S, A, R, O, T, E, L}

- Fields: [0-9]

- Constraints:

      - I != 0, R != 0

      100 * I + 10 * S + A

              +

      100 * R + 10 * O + A

              =

      1000 * T + 100 * E + 10 * L + O
   

```python
var = { "I", "S", "A", "R", "O", "T", "E", "L"}
domain = {i: list(range(10)) for i in var}
solution = {}

domain[0] = list(range(1-10))
domain[1] = list(range(1-10))
```

```python
num = [0] * len(var)

while True:

    for i in range(var):
        solution[var[i]] = num[i]

    if len(set(var.values())) == len(var):
        
        if solution['I'] != 0 and solution['R'] != 0 :

            I = solution['I']; S = solution['S']; A = solution['A']
            R = solution['R']; O = solution['O']; T = solution['T']
            E = solution['E']; L = solution['L']

            ISA = 100 * I + 10 * S + A
            ROA = 100 * R + 10 * O + A
            TELO = 1000 * T + 100 * E + 10 * L + O

            if ISA + ROA == TELO:
                print("Solution trouvee !")
                print(solution)
                print(f"  {I}{S}{A}")
                print(f"+ {R}{O}{A}")
                print(f"= {T}{E}{L}{O}")
                break

    i = len(var) - 1
    while i >= 0:
        num[i] += 1
        if num[i] < 10:
            break
        else:
            num[i] = 0
            i -= 1

    if i < 0:
        print("Pas de solution")
        break
```
