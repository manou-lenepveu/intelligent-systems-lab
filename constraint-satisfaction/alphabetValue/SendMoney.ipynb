### Alphabet Number Problem

Problem: SEND + MORE = MONEY

Variables  : S, E, N, D, M, O, R, Y  (8 lettres)

Domaines   : {0, 1, 2, ..., 9} pour chaque lettre

Contraintes:

  1. Toutes les lettres ont des valeurs différentes (AllDiff)
  
  2. S ≠ 0 et M ≠ 0  (pas de zéro en tête)
  
  3.   SEND + MORE = MONEY
  
       soit :

           1000*S + 100*E + 10*N + D


            + 1000*M + 100*O + 10*R + E


            = 10000*M + 1000*O + 100*N + 10*E + Y

# Generate & Test

```python
lettres = ['S', 'E', 'N', 'D', 'M', 'O', 'R', 'Y']
domain = {l: list(range(10)) for l in lettres}
solution = {}

# Contrainte : S et M ne peuvent pas être 0
domain['S'] = list(range(1, 10))
domain['M'] = list(range(1, 10))
```

```python
n = len(lettres)  

# Assignation initiale : toutes à 0
assign = [0] * n
print("Depart :", assign)

count = 0  # nombre de combinaisons testees
```

```python
while True:

    count += 1

    # GENERATE → indices vers chiffres
    sol = {}
    for i in range(n):
        sol[lettres[i]] = assign[i]

    # TEST 1 : tous les chiffres différents
    if len(set(sol.values())) == n:

        # TEST 2 : S != 0 et M != 0
        if sol['S'] != 0 and sol['M'] != 0:

            # TEST 3 : SEND + MORE = MONEY
            S=sol['S']; E=sol['E']; N=sol['N']; D=sol['D']
            M=sol['M']; O=sol['O']; R=sol['R']; Y=sol['Y']

            SEND  = 1000*S + 100*E + 10*N + D
            MORE  = 1000*M + 100*O + 10*R + E
            MONEY = 10000*M + 1000*O + 100*N + 10*E + Y

            if SEND + MORE == MONEY:
                print("Solution trouvee !")
                print(sol)
                print(f"  {S}{E}{N}{D}")
                print(f"+ {M}{O}{R}{E}")
                print(f"= {M}{O}{N}{E}{Y}")
                print(f"\nCombinaisons testees : {count}")
                break

    # GENERATE : increment comme un compteur en base 10
    i = n - 1
    while i >= 0:
        assign[i] += 1
        if assign[i] < 10:     
            break
        else:
            assign[i] = 0
            i -= 1

    if i < 0:
        print("Pas de solution")
        break
```
